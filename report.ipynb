```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Preprocess :

```python
iris = pd.read_csv('Iris.csv')

enumartedSpecies = {'Iris-setosa' : 0, 'Iris-versicolor' : 1 , 'Iris-virginica' : 2}
iris['Species'] = iris['Species'].map(enumartedSpecies)
print(iris.head())
```

```python
iris_modified = iris.drop("Id",axis=1)
iris_modified = iris_modified.drop("Species",axis=1)
iris_list = iris_modified.values.tolist()
print(iris_list)
```

```python
from sklearn.model_selection import train_test_split
X = iris.iloc[:, 1:-1].values
Y = iris.iloc[:, -1].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
X_train_list = X_train.tolist()
y_train_list = y_train.tolist()
print(X_train_list)
print(y_train_list)
```

Training :

```python
from kmeans import Kmeans
k_means = Kmeans(k=3)
cents, cluster_assignments, num_iter, cents_orig = k_means.fit(X_train_list)
print(cluster_assignments)
#print(y_train_list)
#print(num_iter)
#print(cents)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
target = iris.target

petal_width = data[:, 0]
sepal_length = data[:, 1]
petal_length = data[:, 2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(petal_length, sepal_length, petal_width,c=target, cmap='viridis')

ax.set_zlabel('Petal Width')
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Petal Length')

plt.title("Iris Veri Seti 3D")
plt.show()
```

```python
first_indices_of_cluster_assignments = [[inner_list[0]] for inner_list in cluster_assignments]
print(first_indices_of_cluster_assignments)
print((y_train_list))
```

Predict :

```python
X_test_list = X_test.tolist()
y_pred_list = k_means.predict(X_test_list)
y_test_list = y_test.tolist()
first_indices_of_predicts = [[inner_list[0]] for inner_list in y_pred_list]
print(first_indices_of_predicts)
print(y_test_list)
```

Results :

```python
target3 = first_indices_of_predicts


petal_length_list = []
sepal_length_list = []
petal_width_list =  []

for inner_list in X_test_list:
    petal_length_list.append(inner_list[2])
    sepal_length_list.append(inner_list[0])
    petal_width_list.append(inner_list[3])


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(petal_length_list, sepal_length_list, petal_width_list,c=target3, cmap='viridis')


ax.set_zlabel('Petal Width')
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Petal Length')


plt.title("Test Data'mizin Cluster'lanmis Hali")

plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score,confusion_matrix,roc_curve,auc
flat_y_test_list = [inner_list[0] for inner_list in y_test_list]
flat_y_pred_list = [inner_list[0] for inner_list in y_pred_list]

print("Results of our model :")
confusion_matrix =confusion_matrix(flat_y_test_list,flat_y_pred_list)
print("Confusion Matrix : ")
print(confusion_matrix)

f1 = f1_score(flat_y_test_list,flat_y_pred_list, average='weighted')
print("F1-Score:",f1)

accuracy = accuracy_score(flat_y_test_list,flat_y_pred_list)
print("Accuracy:",accuracy)

precision = precision_score(flat_y_test_list,flat_y_pred_list, average='weighted')
print("Precision:",precision)

recall = recall_score(flat_y_test_list, flat_y_pred_list, average='weighted')
print("Recall:", recall)
```

Burada şöyle bir durum çıktığını belirtmem gerekiyor yoksa yanlış yönlenebiliriz : Clustering label'sız çalıştığından her ne kadar doğru verileri de birbiriyle birleştirse de bunun hangi sınıfta , hangi label'dan olduğunu bilmiyor ve kafasına göre sınıf atıyor. Ve bu durum her bir model denemesinden denemesine farklılık gösterebiliyor çünkü initial noktalar random olarak atanıyor. Bu yüzden burada bu metrikleri kullanarak başarılı veya başarısız olduğumuzu söylememiz çok mantıklı değil. Bunun yerine 3D Plot olarak çizdirdiğimiz grafikte verilerin cluster'lara göre nasıl ayrıldığını görebiliriz. Veya Confusion Matrix'i Actual-Prediction Şeklinde değil de Cluster-Label bazında incelemeliyiz. Örneğin Şu an size yolladığım çıktı için 11 + 13 + 3 veri doğru cluster'lanmışken, 3 veride hata bulunmaktadır. Yani sonuç olarak iyi bir accuracy elde etmiş oluyoruz. Tabii ki belli bir süre denemeden sonra cluster'lar ve label'lar üst üste çakışabilir ve,  ancak o durumda  bu metrikler anlamlı bir sonuç yaratabilir.
Ben yine de çalışan bir durumun ss'lerini ödev dosyasının içine koyacağım.

```python
FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
TP = np.diag(confusion_matrix)
TN = confusion_matrix.sum() - (FP + FN + TP)

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
# Specificity or true negative rate
TNR = TN/(TN+FP) 
# Precision or positive predictive value
PPV = TP/(TP+FP)
# Negative predictive value
NPV = TN/(TN+FN)
# Fall out or false positive rate
FPR = FP/(FP+TN)
# False negative rate
FNR = FN/(TP+FN)
# False discovery rate
FDR = FP/(TP+FP)

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)
print(FPR)
print(TPR)
roc_auc = auc(FPR,TPR)
plt.plot(FPR,TPR, lw=2, label='Class {} (area = {:0.2f})'.format(1, roc_auc))
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

Yine üstte anlattığım durumdan dolayı anlamlı bir grafil elde edemiyoruz. Ama yine de 1 class için de olsa göstermek istedim.

Elbow Method : Which "K" Should We Choose :

```python
sums = []
for i in range(1,10):
    k_meansE = Kmeans(k=i)
    centsE, cluster_assignmentsE, num_iterE, cents_origE = k_meansE.fit(X_train_list)
    second_indices_of_cluster_assignmentsE = [[inner_list[1]] for inner_list in cluster_assignmentsE]
    total_sum = 0
    #print(centsE)
    for inner_list in second_indices_of_cluster_assignmentsE:
        total_sum += inner_list[0]
    sums.append(total_sum)
    
print(sums)
```

```python
plt.scatter(range(1,len(sums) + 1), sums, color='blue', label='SSE')

plt.plot(range(1,len(sums) + 1), sums, color='red', linestyle='-', label='Grafik')

plt.xlabel('X Ekseni (İndisler [0 dan basliyor (0 => k=1)])')
plt.ylabel('Y Ekseni (SSE)')
plt.title('Elbow Method Graphic')

plt.xticks(range(1,len(sums)+1) , range(len(sums)))

plt.legend()

plt.show()
```

Yorum ve Kıyas(DT ile) Kısmı : 

Kmeans ve DT ikisi de kolay anlaşılabilir algoritmalar. Ancak birbirine göre çok temel farklılıkları var:
Öncelikle DT bir supervised-learning algoritması ve training yaparken elimizde çoktan label'lar var. Buna göre iyileştirme ve düzeltme yapabiliyoruz.
Ancak kmeans'te elimizde label yok. Sadece uzaydaki mesafe farkı fonksiyonları kullanarak verisetimizin noktaları arasındaki mesafeleri yorumlayarak ve öbekleyerek belli bir netice alabiliyoruz.
Üstte daha da ayrıntılı anlattığım sıkıntılardan dolayı kmeans her zaman doğru label'ları veremeyebiliyor çünkü aslında arka planda ne yaptığını bilmiyor. Sadece grupluyor. Bu yüzden tam da bir öğrenme diyemeyiz.
Ancak DT doğru yerlerden veriyi bölerek entropiyi en az seviyede tutmaya çalışıyor.
Yine de label'lardan bağımsız düşünürsek kmeans algoritmamız bize kötü bir sonuç vermiyor çünkü aynı label'lara sahip olan verilerimizi aynı öbeğe sokmaya başarıyor. Ancak bu öbek_no ile label_no her zaman aynı olmuyor.
İki algoritmanın da zayıf yönleri de var.
Örneğin DT'de dikkat edilmesi gereken en önemli unsurlardan biri de fit ederken verdiğimiz hiperparametreler.Mesela ağacın uzunluğu. Bu parametrelere dikkat edilmez ise overfit olabilir.
Kmeans'te de buna benzer olarak k sayısı var. Eğer yeterince fazla k koyarsak tüm verileri teker teker farklı öbeklere koyabiliriz ve yine bu istediğimiz bir durum değil. Bunun için SSE kullanarak Elbow method ile 1-10 arası k değerleri için modeli denedik.İlk gruplamanın başladığı k=2 olarak düşünürsek veri setimiz için label sayımız olan 3 değeri için modelin başarısı diğer durumlara göre oransal olarak daha fazla arttı.(1-2 geçişini saymıyorum çünkü 1 de bir clustering yok diyebiliriz bu yüzden bu geçişteki oran diğerlerine göre çok çok fazla olacaktır.)3'den sonra bu oran giderek düşmeye devam etti.
Burada tabii ki veri setimizin kmeans kullanılabilecek bir veri seti olması da işimize yaradı. Örneğin iki veya daha fazla hiperbolik alanda değişkenlik gösteren veri noktalarını gruplandırmamız istenseydi, bizim yazdığımız algoritma bunu farklı şekilde yapardı çünkü merkez tanımımız bir nokta.
